```
# Function to validate a mobile recharge request
def validate_recharge(amount):
    # List of valid recharge plans in rupees
    valid_plans = [199, 299, 399, 599]

    # Rule 1: Recharge amount must be at least Rs.50
    if amount < 50:
        return False, "Amount must be >= Rs.50"

    # Rule 2: Amount must match one of the valid plans
    if amount not in valid_plans:
        return False, f"Invalid plan. Choose from: {valid_plans}"

    # If both rules pass, recharge is valid
    return True, "Recharge Successful!"


# Test cases: includes invalid amount, invalid plan, and valid plans
test_inputs = [30, 150, 199, 299, 399, 599]

# Loop through each test input and display the validation result
for amt in test_inputs:
    status, msg = validate_recharge(amt)
    print(f"Amount Rs.{amt}: {msg}")
```

```
# Function to check stock levels and print appropriate alerts
def check_inventory(inventory):
    # Iterate over each product and its stock quantity in the dictionary
    for product, stock in inventory.items():
        # Trigger reorder alert if stock falls below threshold (15 units)
        if stock < 15:
            print(f"{product}: Reorder Alert (Stock: {stock})")
        else:
            # Stock is sufficient, no action needed
            print(f"{product}: Stock OK (Stock: {stock})")


# Sample inventory: product name mapped to current stock quantity
inventory = {
    "Laptop": 5,       # Below threshold -> Reorder Alert
    "Mouse": 20,       # Above threshold -> Stock OK
    "Keyboard": 12,    # Below threshold -> Reorder Alert
    "Monitor": 8,      # Below threshold -> Reorder Alert
    "Earpods": 30   # Above threshold -> Stock OK
}

# Call the function to check and display inventory status
check_inventory(inventory)
```

```
# Function to process student marks and determine Pass/Fail
def process_results(marks):
    total = 0

    # Calculate total marks using a loop
    for m in marks:
        total += m

    # Compute average by dividing total by number of subjects
    average = total / len(marks)

    # Determine result: Pass if average >= 50, else Fail
    result = "Pass" if average >= 50 else "Fail"

    # Return both the average and the result
    return average, result


# Dictionary of students and their subject-wise marks
students = {
    "Arun": [78, 85, 90, 72, 88],  # High scorer -> Pass
    "Sanjay":   [40, 35, 48, 50, 42],  # Low scorer  -> Fail
    "Charlie": [60, 55, 70, 65, 58]   # Average     -> Pass
}

# Process and display result for each student
for name, marks in students.items():
    avg, result = process_results(marks)
    print(f"{name}: Average = {avg:.1f} => {result}")
```

```
# Function to calculate cab fare based on distance and time of travel
def calculate_fare(distance, peak_hour):
    base_fare = 50          # Fixed base fare in rupees
    per_km_rate = 12        # Charge per kilometer

    # Calculate fare: base fare + per km charge
    fare = base_fare + (per_km_rate * distance)

    # Apply 25% surcharge during peak hours
    if peak_hour:
        fare *= 1.25

    return fare


# Test cases: (distance in km, is_peak_hour)
test_cases = [
    (10, False),   # 10 km, normal hour
    (10, True),    # 10 km, peak hour
    (25, False),   # 25 km, normal hour
    (25, True)     # 25 km, peak hour
]

# Loop through each test case and display the estimated fare
for distance, peak in test_cases:
    fare = calculate_fare(distance, peak)
    peak_label = "Peak Hour" if peak else "Normal Hour"
    print(f"Distance: {distance} km | {peak_label} | Fare: Rs.{fare:.2f}")
```

```
# Function to calculate attendance percentage and determine eligibility
def check_attendance(attendance):
    present = 0

    # Count the number of days marked as Present ('P')
    for day in attendance:
        if day == "P":
            present += 1

    # Calculate attendance percentage
    percentage = (present / len(attendance)) * 100

    # Employee is eligible only if attendance is 75% or above
    status = "Eligible" if percentage >= 75 else "Not Eligible"

    # Return both the percentage and eligibility status
    return percentage, status


# Attendance records: 'P' = Present, 'A' = Absent
employees = {
    "Arun":  ["P", "P", "A", "P", "P", "A", "P", "P", "P", "P"],  # 80% -> Eligible
    "Meerakshi": ["P", "A", "A", "P", "A", "P", "A", "P", "A", "A"],  # 40% -> Not Eligible
    "Kavitha": ["P", "P", "P", "P", "P", "P", "P", "P", "A", "P"]   # 90% -> Eligible
}

# Check and display attendance eligibility for each employee
for emp, att in employees.items():
    pct, status = check_attendance(att)
    print(f"{emp}: Attendance = {pct:.1f}% => {status}")
```

```
# Function to evaluate the strength of a given password
def check_password_strength(password):
    special_chars = "@#$"   # Allowed special characters for validation
    has_digit = False        # Flag to track if a digit is found
    has_special = False      # Flag to track if a special character is found

    # Rule 1: Password must be at least 8 characters long
    if len(password) < 8:
        return "Weak: Password too short (min 8 characters)"

    # Scan each character in the password to check for digit and special char
    for ch in password:
        if ch.isdigit():        # Check if character is a digit
            has_digit = True
        if ch in special_chars: # Check if character is a special character
            has_special = True

    # Rule 2: Password must contain at least one digit
    if not has_digit:
        return "Weak: Must contain at least one digit"

    # Rule 3: Password must contain at least one special character
    if not has_special:
        return "Weak: Must contain at least one special character (@, #, $)"

    # All rules passed -> Strong password
    return "Strong Password"


# Test passwords covering all validation scenarios
passwords = [
    "abc",       # Too short
    "abcdefgh",  # No digit
    "abcdefg1",  # No special character
    "abcd@123",  # All rules satisfied -> Strong
    "Secure@9"   # All rules satisfied -> Strong
]

# Evaluate and display strength for each password
for pwd in passwords:
    result = check_password_strength(pwd)
    print(f"Password: '{pwd}' => {result}")
```
