```
# Problem 1: Unique Words in a Sentence
# Goal: Find all unique words from a sentence and display their count

# Take input from the user
sentence = input("Enter sentence: ")

# Split the sentence into individual words
words = sentence.split()

# Convert the list to a set to automatically remove duplicates
unique_words = set(words)

# Display the count and the unique words
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```
# Problem 2: Highest Salary from Employee Data
# Goal: Find the employee with the highest salary from a dictionary

# Dictionary with employee names as keys and salaries as values
employees = {
    "Arvind": 175000,
    "Kavitha": 68000,
    "Arun Kumar": 72000
}

# Use max() with key argument to find the employee with the highest salary
highest_paid = max(employees, key=employees.get)

# Display the result
print("Highest Salary:", highest_paid, "-", employees[highest_paid])
```

```
# Problem 3: Find Maximum and Minimum Values
# Goal: Find the max and min from a list using a loop (without built-in max/min)

numbers = [54, 33, 98, 10, 55]

# Initialize max and min with the first element of the list
maximum = numbers[0]
minimum = numbers[0]

# Loop through the list and compare each element
for num in numbers:
    if num > maximum:
        maximum = num   # Update max if current number is greater
    if num < minimum:
        minimum = num   # Update min if current number is smaller

# Display the results
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```
# Problem 4: Count Products Above a Price Threshold
# Goal: Count how many products have a price greater than 1000

prices = [550, 1300, 699, 1800, 400]

# Initialize a counter
count = 0

# Loop through each price and compare with 1000
for price in prices:
    if price > 1000:
        count += 1   # Increment counter if price exceeds threshold

# Display the result
print("Products above 1000:", count)
```

```
# Problem 5: Calculate Attendance Percentage
# Goal: Calculate the percentage of days a student was present

attendance = ["P", "P", "A", "P", "P","A"]

# Count the number of 'P' (Present) entries
present_days = attendance.count("P")

# Total number of days
total_days = len(attendance)

# Calculate attendance percentage
attendance_percentage = (present_days / total_days) * 100

# Display the result
print("Attendance Percentage:", attendance_percentage)
```

```
# Problem 6: Remove Duplicate Phone Numbers
# Goal: Remove duplicates from a list of phone numbers using a set

phone_numbers = [9876543210, 9123456789, 9876543210]

# Convert the list to a set — sets automatically remove duplicates
unique_phone_numbers = set(phone_numbers)

# Display the unique phone numbers
print("Unique phone numbers:", unique_phone_numbers)
```

```
# Problem 7: Count Character Frequency
# Goal: Count how many times each character appears in a string

word = "python"

# Use an empty dictionary to store character counts
char_frequency = {}

# Loop through each character in the string
for char in word:
    if char in char_frequency:
        char_frequency[char] += 1   # Character already seen — increment count
    else:
        char_frequency[char] = 1    # First occurrence — set count to 1

# Display the frequency dictionary
print(char_frequency)
```

```
# Problem 8: Convert List to Tuple
# Goal: Convert a list into a tuple (immutable/read-only structure)

my_list = [40, 50, 60]

# Use the tuple() function to convert the list
my_tuple = tuple(my_list)

# Display the tuple
print("Tuple:", my_tuple)
```

```
# Problem 9: Check if a Key Exists in a Dictionary
# Goal: Verify whether a specific key is present in a dictionary

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Key to search for
search_key = "Ravi"

# Use the 'in' keyword to check if the key exists
if search_key in employees:
    print("Employee exists")
else:
    print("Employee not found")
```

```
# Problem 10: Calculate Average Marks
# Goal: Compute the average of a student's marks

marks = [70, 85, 60, 75]  # List of marks (average = 72.5)

# Calculate the sum of all marks
total = sum(marks)

# Divide by the total number of subjects to get the average
average = total / len(marks)

# Display the result
print("Average Marks:", average)
```
