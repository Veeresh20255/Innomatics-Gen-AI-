```
import re

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return []

    # Lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove emails
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters (soooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove punctuation except spaces
    text = re.sub(r'[^\w\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short tokens (≤2) except "no", "not"
    cleaned_tokens = [
        word for word in tokens
        if len(word) > 2 or word in ["no", "not"]
    ]

    return cleaned_tokens

print(preprocess_text("Get 100% FREE access now!!!"))
```

```
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

def display_results(sentences):
    all_tokens = []

    for text in sentences:
        tokens = preprocess_text(text)
        clean_sentence = " ".join(tokens)

        all_tokens.extend(tokens)

        print("\n---")
        print("Original Text:", text)
        print("Cleaned Tokens:", tokens)
        print("Cleaned Sentence:", clean_sentence)

        # Analytics
        if tokens:
            total = len(tokens)
            unique = len(set(tokens))
            avg_len = sum(len(w) for w in tokens) / total
        else:
            total = unique = avg_len = 0

        print("Total Tokens:", total)
        print("Unique Tokens:", unique)
        print("Avg Token Length:", round(avg_len, 2))

    return all_tokens

all_tokens = display_results(test_sentences)
```

```
import re

# --- Task 1 ---
def preprocess_text(text):
    if not text or not isinstance(text, str):
        return []

    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()
    tokens = [w for w in tokens if len(w) > 2 or w in ["no", "not"]]

    return tokens


# --- Task 4 ---
def token_analytics(sentences):
    for text in sentences:
        tokens = preprocess_text(text)

        total_tokens = len(tokens)
        unique_tokens = len(set(tokens))
        avg_length = round(sum(len(word) for word in tokens) / total_tokens, 2) if total_tokens > 0 else 0

        print("\n---")
        print("Original Text:", text)
        print("Tokens:", tokens)
        print("Cleaned Sentence:", " ".join(tokens))
        print("Total Tokens:", total_tokens)
        print("Unique Tokens:", unique_tokens)
        print("Avg Token Length:", avg_length)


# 🔥 THIS IS WHAT YOU ARE MISSING
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍"
]

token_analytics(test_sentences)
```

```
def full_pipeline(text_list):
    if not isinstance(text_list, list):
        raise ValueError("Input must be a list of strings")

    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)
        sentence = " ".join(tokens)

        all_tokens.extend(tokens)
        clean_sentences.append(sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
    test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

result = full_pipeline(test_sentences)

print(result)
```
